```python
import os
import sys
sys.path.insert(0,r'/Users/seanippolito/repos/theseankid/computational-topology/distances')
import ellipsoid_distances
import tadasets
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse, Circle

import sympy as sym
from sympy import lambdify
from sympy import Matrix

x = sym.Symbol('x')
y = sym.Symbol('y')
f = x**2 + y**2 - 1
variables = [x,y]
```

```python
seed=5
data=tadasets.dsphere(n=50, d=1)
sigma = 0.2

Vs = ellipsoid_distances.jacobian_rotation_matricies(data,f,variables)
SVDs, _ = ellipsoid_distances.svd_rotation_matricies(data, sigma = 0.2, nnbrs=1,fixed = True)
```

```python
fig2,ax2=plt.subplots(nrows=4, ncols=3, figsize=(15,20))


eps = 0.1
ax2[0,0].scatter(data[:,0], data[:,1], s=3)
for row in data:
    e = Circle(xy=row, radius=eps, alpha = 0.25)
    ax2[0,0].add_patch(e)
    
ax2[0,1].scatter(data[:,0],data[:,1], s=3)
ax2[0,2].scatter(data[:,0],data[:,1], s=3)


for row, V, SVD in zip(data, Vs, SVDs):
    e = Ellipse(xy=row, width=2*eps, height=2*sigma*eps, angle=np.degrees(np.math.acos(V[1,0])),
        alpha = 0.25 )
    ax2[0,1].add_patch(e)
    
    e = Ellipse(xy=row, width=2*sigma*eps, height=2*eps, angle=np.degrees(np.math.acos(SVD[1,1])),
        alpha = 0.25 )
    ax2[0,2].add_patch(e)
    
ax2[0,0].title.set_text('$\epsilon (0.1)$ Balls')
ax2[0,1].title.set_text('Jacobian Ellipsoids: $\epsilon = 0.1$')
ax2[0,2].title.set_text('SVD Ellipsoids: $\epsilon = 0.1$')


eps = 0.2
ax2[1,0].scatter(data[:,0], data[:,1], s=3)
for row in data:
    e = Circle(xy=row, radius=eps, alpha = 0.25)
    ax2[1,0].add_patch(e)
    
ax2[1,1].scatter(data[:,0],data[:,1], s=3)
ax2[1,2].scatter(data[:,0],data[:,1], s=3)

for row, V, SVD in zip(data, Vs, SVDs):
    e = Ellipse(xy=row, width=2*eps, height=2*sigma*eps, angle=np.degrees(np.math.acos(V[1,0])),
        alpha = 0.25 )
    ax2[1,1].add_patch(e)
    
    e = Ellipse(xy=row, width=2*sigma*eps, height=2*eps, angle=np.degrees(np.math.acos(SVD[1,1])),
        alpha = 0.25 )
    ax2[1,2].add_patch(e)
    
ax2[1,0].title.set_text('$\epsilon (0.2)$ Balls')
ax2[1,1].title.set_text('Jacobian Ellipsoids: $\epsilon = 0.2$')
ax2[1,2].title.set_text('SVD Ellipsoids: $\epsilon = 0.2$')


eps = 0.3
ax2[2,0].scatter(data[:,0], data[:,1], s=3)
for row in data:
    e = Circle(xy=row, radius=eps, alpha = 0.25)
    ax2[2,0].add_patch(e)
    
ax2[2,1].scatter(data[:,0],data[:,1], s=3)
ax2[2,2].scatter(data[:,0],data[:,1], s=3)


for row, V, SVD in zip(data, Vs, SVDs):
    e = Ellipse(xy=row, width=2*eps, height=2*sigma*eps, angle=np.degrees(np.math.acos(V[1,0])),
        alpha = 0.25 )
    ax2[2,1].add_patch(e)
    
    e = Ellipse(xy=row, width=2*sigma*eps, height=2*eps, angle=np.degrees(np.math.acos(SVD[1,1])),
        alpha = 0.25 )
    ax2[2,2].add_patch(e)
    
ax2[2,0].title.set_text('$\epsilon (0.3)$ Balls')
ax2[2,1].title.set_text('Jacobian Ellipsoids: $\epsilon = 0.3$')
ax2[2,2].title.set_text('SVD Ellipsoids: $\epsilon = 0.3$')


eps = 0.4
ax2[3,0].scatter(data[:,0], data[:,1], s=3)
for row in data:
    e = Circle(xy=row, radius=eps, alpha = 0.25)
    ax2[3,0].add_patch(e)
    
ax2[3,1].scatter(data[:,0],data[:,1], s=3)
ax2[3,2].scatter(data[:,0],data[:,1], s=3)

for row, V, SVD in zip(data, Vs, SVDs):
    e = Ellipse(xy=row, width=2*eps, height=2*sigma*eps, angle=np.degrees(np.math.acos(V[1,0])),
        alpha = 0.25 )
    ax2[3,1].add_patch(e)
    
    e = Ellipse(xy=row, width=2*sigma*eps, height=2*eps, angle=np.degrees(np.math.acos(SVD[1,1])),
        alpha = 0.25 )
    ax2[3,2].add_patch(e)
    
ax2[3,0].title.set_text('$\epsilon (0.4)$ Balls')
ax2[3,1].title.set_text('Jacobian Ellipsoids: $\epsilon = 0.4$')
ax2[3,2].title.set_text('SVD Ellipsoids: $\epsilon = 0.4$')

fig2.savefig('figures_01-12.png')
```
